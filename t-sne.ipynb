```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
#import umap
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

df = pd.read_csv('D://train_data.csv',index_col=0)
exclude = ["no2","formula","r1","r2",'tag']
X = df.drop(exclude,axis=1)
print(X.shape)
```

```python
scaler = MinMaxScaler()
scaler = scaler.fit(X)
X_m = scaler.transform(X)
X_embedding = TSNE(n_components=2, learning_rate='auto',init='random', perplexity=3,random_state=0).fit_transform(X_m)
#print(X_embedding)
#
# reducer = umap.UMAP(random_state=18)  ## 13 12
# X_embedding = reducer.fit_transform(X_m)
plt.scatter(X_embedding[:, 0], X_embedding[:, 1], c=df.tag, cmap='Spectral', s=5)
plt.gca().set_aspect('equal', 'datalim')
#plt.colorbar(boundaries=np.arange(11)-0.5).set_ticks(np.arange(10))
plt.title('t-SNE projection of the Digits dataset')
plt.show()

print(X_embedding.shape)
pd_reducerdata = pd.DataFrame(X_embedding)
pd_reducerdata.to_csv('D://X_embedding_tsne.csv')
```
